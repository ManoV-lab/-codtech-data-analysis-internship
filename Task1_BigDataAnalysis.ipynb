```python
from pyspark.sql import SparkSession

# Create or get a Spark Session
spark = SparkSession.builder.appName("BigDataAnalysis").getOrCreate()

print("Spark session created successfully!")
```

```python
# Load a sample CSV file from Colab's builtâ€‘in data folder
df = spark.read.csv("/content/sample_data/mnist_test.csv", header=True, inferSchema=True)

# Show first 5 rows
df.show(5)

# Display the schema (data types of each column)
df.printSchema()
```

```python
# Count how many rows (records) are in the dataset
print("Number of rows:", df.count())

# Describe numeric columns (mean, min, max, etc.)
df.describe().show()
```

```python
# Pick two columns from the dataset
cols = df.columns
if len(cols) >= 2:
    # Group by second column, calculate average of the first
    df.groupBy(cols[1]).avg(cols[0]).show()
else:
    print("Dataset too small for grouping.")
```

```python
df.coalesce(1).write.csv("/content/output_summary.csv", header=True)
```

- The operations ran in parallel across columns efficiently.
